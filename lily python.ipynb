# FILLING MISSING VALUES
1. Unknown
2. most frequent

```python
df.fillna(value={"Manufacturer":"unknown"}, inplace=True) 
#the inplace function changes values directly into the original dataset
```

```python
df.head(10)
```

```python
average=df['Sales'].mean() #we want to get the average and therefore we replace missing value with 'sales'
df.fillna(value={'Sales':average},inplace=True)
```

```python
df.head(10)
```

```python
df.dropna(inplace=True) #dropna removes column with nan
```

```python
df.head(10)
```

```python
df[df.duplicated()] #shows/returns duplicatesexisting
```

```python
df.drop_duplicates(inplace=True) # rermove duplicates in rows
```

# DETECTING OUTLIERS

```python
df.head()
```

```python
q1=df["Sales"].quantile(0.25)
q3=df["Sales"].quantile(0.75)
iqr=q3-q1
print(q1,q3,iqr)
```

```python
lower_bond=q1-1.5*iqr
```

```python
lower_bond
```

```python
upper_bond=q3+1.5*iqr
```

```python
upper_bond
```

```python
required=df[(df["Sales"]>=lower_bond) &(df["Sales"]<=upper_bond)]
print("all Sales\n",df["Sales"].values)
print("Sales after removing outliers\n",required["Sales"].values)
```
